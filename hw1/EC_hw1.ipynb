```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from copy import deepcopy
```

```python
# constant declarations
population_size, length = 200, 50
shift = 0
runs = 10
generation = 100
```

## 3. Roulette wheel & (fitness = sum)

```python
# 10 runs
shift = 0
best_per_run = []
for r in range(runs):
    
    # random init
    chromo = np.random.randint(2,  size=(population_size, length))
    children = np.zeros((population_size, length))
    np.random.shuffle(chromo)

    best_per_generation = []
    fitness = np.sum(chromo, axis=1) + shift

    # 100 generations
    for g in range(generation):

        np.random.shuffle(chromo)

        # parent selection
        # Roulette wheel selection
        propotion = fitness/sum(fitness)


        for i in range(population_size):
            selected_parent = np.random.choice(population_size, 1, p=propotion)[0]
            #print(selected_parent)
            children[i] = chromo[selected_parent].copy()

        # crossover
        for i in range(population_size):
            #flag_crossover = np.random.choice(2, 1, p=[0.9, 0.1])[0]
            flag_crossover = 1
            if(flag_crossover == 1):
                cxpoint = np.random.choice(length-1, 1)[0]+1
                i1 = (i+1) % population_size
                children[i][cxpoint:length], children[i1][cxpoint:length] = \
                    children[i1][cxpoint:length].copy(), children[i][cxpoint:length].copy() 
            #best_per_generation.append(prob_crossover)

        # survivor selection (generational)
        chromo = deepcopy(children)

        fitness = np.sum(chromo, axis=1) + shift
        best_per_generation.append(max(fitness - shift))
        
    best_per_run.append(best_per_generation)


plt.title('3. Roulette wheel & fitness = sum')
plt.xlabel('generation')
plt.ylabel('average best fitness value')
plt.plot(np.mean(best_per_run, axis=0))
plt.show()
```

## 6. tournament & (fitness = sum)

```python
# 10 runs
best_per_run = []
for r in range(runs):
    
    # random init
    chromo = np.random.randint(2,  size=(population_size, length))
    children = np.zeros((population_size, length))
    np.random.shuffle(chromo)

    best_per_generation = []
    fitness = np.sum(chromo, axis=1) + shift
    
    for g in range(generation):
        np.random.shuffle(chromo)

        # parent selection
        # Roulette wheel selection

        k = 2

        for i in range(population_size):
            i1 = (i+1) % population_size
            selected_parent = [i, i1]
            win = np.argmax(fitness[selected_parent])
            win_index = selected_parent[win]
            children[i] = chromo[win_index].copy()
     
            
        # crossover
        for i in range(population_size):
            #flag_crossover = np.random.choice(2, 1, p=[0.9, 0.1])[0]
            flag_crossover = 1
            if(flag_crossover == 1):
                cxpoint = np.random.choice(length-1, 1)[0]+1
                i1 = (i+1) % population_size
                children[i][cxpoint:length], children[i1][cxpoint:length] = \
                    children[i1][cxpoint:length].copy(), children[i][cxpoint:length].copy() 
            #best_per_generation.append(prob_crossover)

        # survivor selection (generational)
        chromo = deepcopy(children)
        fitness = np.sum(chromo, axis=1) + shift
        best_per_generation.append(max(fitness - shift))

    best_per_run.append(best_per_generation)

    
plt.title('6. tournament & (fitness = sum)')
plt.xlabel('generation')
plt.ylabel('average best fitness value')
plt.plot(np.mean(best_per_run, axis=0))
plt.show()
```
